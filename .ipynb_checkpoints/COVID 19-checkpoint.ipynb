```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('WHO-COVID-19-global-data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.shape
```

```python
country_cases = df.groupby('Country')['Cumulative_cases'].max().sort_values(ascending=False)
```

```python
country_cases
```

```python
plt.figure(figsize=(12, 8))
country_cases.plot(kind='bar', color='skyblue')
plt.title('Total COVID-19 Cases by Country')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.show()
```
